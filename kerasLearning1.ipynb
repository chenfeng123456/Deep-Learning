```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import math

num_points = 10000
vectors_set = []
x_data = []
y_data = []
x_data2 = []
y_data2 = []
for i in range(num_points):
    x1 = np.random.normal(0.0, 0.55) * 2.5
    y1 = np.sin(x1) + np.random.normal(0.0, 0.03) # 避免问题太简单，增加点难度
    vectors_set.append([x1, y1])
    if i <8000:
        x_data.append(x1)
        y_data.append(y1)
    else:
        x_data2.append(x1)
        y_data2.append(y1)

        
x_train = np.array([x_data])
y_train = np.array([y_data])
x_test = np.array([x_data2])
y_test = np.array([y_data2])

print(x_train.shape)
print(y_train.shape)
```

```python
from keras.models import Sequential

model = Sequential()
```

```python
from keras.layers import Dense, Activation

model.add(Dense(units = 64, input_dim = 8000))
model.add(Activation("relu"))
model.add(Dense(8000))
model.add(Activation("softmax"))
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'sgd', metrics = ['accuracy'])
```

```python
from keras.optimizers import SGD
model.compile(loss = 'categorical_crossentropy', optimizer = SGD(lr = 0.01, momentum = 0.9, nesterov = True))
```

```python
model.fit(x_train, y_train, epochs = 5000, batch_size = 32)
```

```python
model.train_on_batch(x_train, y_train)
```

```python
loss_and_metrics = model.evaluate(x_train, y_train, batch_size = 128)
```

```python
classes = model.predict(x_train, batch_size = 128)
```
