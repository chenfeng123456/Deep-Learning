## 1.3 颜色空间转换
在OpenCV中，图像不是用常规的RGB颜色通道来存储的，他们用的顺序是BGR顺序。当读取一幅图像后，默认的是BGR，不过有很多转换方式可以利用。颜色空间转换函数可以用cvtColor()完成。

```python
import cv2
```

```python
# 载入图像
im = cv2.imread('./0.png')
print(im.shape)
```

```python
# creat a grayscale version
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
print(gray.shape)
```

```python
im = cv2.imread('0.png')

gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

# 显示原始图像
fig = plt.figure()
subplot(121)
plt.gray()
imshow(im)
title(u'color image')
axis('off')

# 显示灰度化图像
plt.subplot(122)
plt.gray()
imshow(gray)
title(u'gray image')
axis('off')

show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

red = np.copy(im[:, :, :])
red[:, :, 1] = 0
red[:, :, 2] = 0
plt.imshow(red)
```

```python
import cv2
import numpy as np

# 添加椒盐噪声
for k in range(100):
    # 建立图片随机坐标点
    i = int(np.random.random() * im.shape[0])
    j = int(np.random.random() * im.shape[1])
    # 图片为灰度图像， 有二维
    if im.ndim == 2:
        im[i, j] = 255
    # 图片为彩色图片，有三维，RGB
    elif im.ndim == 3:
        im[i, j, 0] = 255
        im[i, j, 1] = 255
        im[i, j, 2] = 255
        
plt.imshow(im)
```

彩色图片各通道分离与合并 

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('0.png')
# split 返回BGR三个通道
b, g, r = cv2.split(img)

plt.imshow(b)
```

```python
# merged 通道合并
merge = cv2.merge([b, g, r])

# 创建窗口，cv2.namedWindow(窗口名)
cv2.namedWindow('image')
 
# 保存图片,cv2.imwrite(保存图片名,要保存的图片)
cv2.imwrite('0.jpg',img)
# 第三个参数针对特定的格式： 对于JPEG，其表示的是图像的质量，用0-100的整数表示(越高越清晰，压缩级别越低)，默认为95。  注意，cv2.IMWRITE_JPEG_QUALITY类型为Long，必须转换成int。
# cv2.imwrite('003.jpg',img,[int(cv2.IMWRITE_JPEG_QUALITY), 10])
# 对于PNG，第三个参数表示的是压缩级别。cv2.IMWRITE_PNG_COMPRESSION，从0到9,压缩级别越高，图像尺寸越小。默认级别为3
# cv2.imwrite('004.png', img, [int(cv2.IMWRITE_PNG_COMPRESSION), 5])
# 图片显示，cv2.imshow(窗口名，要显示的图片)
cv2.imshow('merge', merge)

# 键盘绑定函数
cv2.waitKey(0)
# 释放窗口
cv2.destroyAllWindows()
```
